### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Cálculo do faturamento total de cada loja

faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print("Faturamento Loja 1: R$ {:.2f}".format(faturamento_loja1))
print("Faturamento Loja 2: R$ {:.2f}".format(faturamento_loja2))
print("Faturamento Loja 3: R$ {:.2f}".format(faturamento_loja3))
print("Faturamento Loja 4: R$ {:.2f}".format(faturamento_loja4))
```

# 2. Vendas por Categoria



```python
# Vendas por categoria em cada loja

vendas_categoria_loja1 = loja['Categoria do Produto'].value_counts()
vendas_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

print("Vendas por categoria - Loja 1:\n", vendas_categoria_loja1, "\n")
print("Vendas por categoria - Loja 2:\n", vendas_categoria_loja2, "\n")
print("Vendas por categoria - Loja 3:\n", vendas_categoria_loja3, "\n")
print("Vendas por categoria - Loja 4:\n", vendas_categoria_loja4)
```

# 3. Média de Avaliação das Lojas

```python
# Média de avaliação das compras em cada loja

media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print("Média de avaliação - Loja 1: {:.2f}".format(media_avaliacao_loja1))
print("Média de avaliação - Loja 2: {:.2f}".format(media_avaliacao_loja2))
print("Média de avaliação - Loja 3: {:.2f}".format(media_avaliacao_loja3))
print("Média de avaliação - Loja 4: {:.2f}".format(media_avaliacao_loja4))
```

# 4. Produtos Mais e Menos Vendidos

```python
# Produtos mais vendidos e menos vendidos em cada loja

print("Produtos mais vendidos - Loja 1:\n", loja['Produto'].value_counts().head(), "\n")
print("Produtos menos vendidos - Loja 1:\n", loja['Produto'].value_counts().tail(), "\n")

print("Produtos mais vendidos - Loja 2:\n", loja2['Produto'].value_counts().head(), "\n")
print("Produtos menos vendidos - Loja 2:\n", loja2['Produto'].value_counts().tail(), "\n")

print("Produtos mais vendidos - Loja 3:\n", loja3['Produto'].value_counts().head(), "\n")
print("Produtos menos vendidos - Loja 3:\n", loja3['Produto'].value_counts().tail(), "\n")

print("Produtos mais vendidos - Loja 4:\n", loja4['Produto'].value_counts().head(), "\n")
print("Produtos menos vendidos - Loja 4:\n", loja4['Produto'].value_counts().tail())
```

# 5. Frete Médio por Loja


```python
# Cálculo do frete médio de cada loja

frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

print("Frete médio - Loja 1: R$ {:.2f}".format(frete_medio_loja1))
print("Frete médio - Loja 2: R$ {:.2f}".format(frete_medio_loja2))
print("Frete médio - Loja 3: R$ {:.2f}".format(frete_medio_loja3))
print("Frete médio - Loja 4: R$ {:.2f}".format(frete_medio_loja4))
```

# 6. Relatório

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Listas das lojas e rótulos
labels_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
dfs = [loja, loja2, loja3, loja4]

# Garantindo que a coluna de data está no formato datetime e criando coluna 'AnoMes'
for df in dfs:
    df['Data da Compra'] = pd.to_datetime(df['Data da Compra'], dayfirst=True, errors='coerce')
    df['AnoMes'] = df['Data da Compra'].dt.to_period('M')

# Faturamento total, avaliação média, frete médio e análise de sazonalidade (desvio padrão mensal)
faturamentos = []
avaliacoes = []
fretes = []
sazonalidade = []
faturamento_mensal = []

for df in dfs:
    faturamentos.append(df['Preço'].sum())
    avaliacoes.append(df['Avaliação da compra'].mean())
    fretes.append(df['Frete'].mean())
    mensal = df.groupby('AnoMes')['Preço'].sum()
    faturamento_mensal.append(mensal)
    sazonalidade.append(mensal.std())

# Normalizando as métricas (quanto maior, melhor para faturamento, avaliação e estabilidade; quanto menor, melhor para frete e sazonalidade)
scaler = MinMaxScaler()

faturamento_norm = scaler.fit_transform(np.array(faturamentos).reshape(-1,1)).flatten()
avaliacao_norm = scaler.fit_transform(np.array(avaliacoes).reshape(-1,1)).flatten()
frete_norm = 1 - scaler.fit_transform(np.array(fretes).reshape(-1,1)).flatten()
estabilidade_norm = 1 - scaler.fit_transform(np.array(sazonalidade).reshape(-1,1)).flatten()  # quanto menor o desvio padrão, maior a estabilidade

# Pesos para cada critério
peso_faturamento = 0.45
peso_avaliacao = 0.25
peso_frete = 0.15
peso_estabilidade = 0.15

# Cálculo da pontuação ponderada final
pontuacao = (
    faturamento_norm * peso_faturamento +
    avaliacao_norm * peso_avaliacao +
    frete_norm * peso_frete +
    estabilidade_norm * peso_estabilidade
)

idx_vender = np.argmin(pontuacao)
loja_para_vender = labels_lojas[idx_vender]
idx_mantidas = [i for i in range(4) if i != idx_vender]

# RELATÓRIO FINAL
print("# Relatório Final – Análise Integrada das Lojas da Alura Store\n")
print("Prezado Sr. João,\n")
print("Realizamos uma análise detalhada dos indicadores de desempenho das quatro lojas da Alura Store, incluindo faturamento, avaliação dos clientes, custo médio de frete e estabilidade ao longo do tempo (sazonalidade).")
print("A seguir, apresentamos os principais resultados e gráficos, bem como a recomendação fundamentada para a decisão de venda.\n")

print("## Principais Indicadores:")
for i in range(4):
    print(f"{labels_lojas[i]}:")
    print(f"  - Faturamento total: R$ {faturamentos[i]:,.2f}")
    print(f"  - Média de avaliação: {avaliacoes[i]:.2f}")
    print(f"  - Frete médio: R$ {fretes[i]:.2f}")
    print(f"  - Estabilidade (desvio padrão do faturamento mensal): {sazonalidade[i]:.2f}")
    print(f"  - Pontuação integrada: {pontuacao[i]:.3f}\n")

# GRÁFICO 1 – Faturamento total por loja (barras)
plt.figure(figsize=(7,5))
plt.bar(labels_lojas, faturamentos, color=['#4285F4', '#EA4335', '#FBBC05', '#34A853'])
plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento (R$)")
plt.xlabel("Loja")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# GRÁFICO 2 – Faturamento mensal (linha) para análise de sazonalidade
plt.figure(figsize=(10,6))
for i, mensal in enumerate(faturamento_mensal):
    plt.plot(mensal.index.astype(str), mensal.values, label=labels_lojas[i])
plt.title("Faturamento Mensal por Loja (Sazonalidade)")
plt.xlabel("Ano-Mês")
plt.ylabel("Faturamento (R$)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# GRÁFICO 3 – Distribuição das avaliações (boxplot)
plt.figure(figsize=(8,6))
plt.boxplot([df['Avaliação da compra'] for df in dfs], labels=labels_lojas, patch_artist=True)
plt.title("Distribuição das Avaliações por Loja")
plt.ylabel("Avaliação da Compra")
plt.tight_layout()
plt.show()

# GRÁFICO 4 – Participação das categorias mais vendidas (pizza) da loja com maior faturamento
idx_maior_faturamento = np.argmax(faturamentos)
df_maior_faturamento = dfs[idx_maior_faturamento]
categorias = df_maior_faturamento['Categoria do Produto'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title(f"Participação das Categorias - {labels_lojas[idx_maior_faturamento]}")
plt.tight_layout()
plt.show()

# TEXTO FINAL E RECOMENDAÇÃO
print("## Considerações e Decisão Recomendada:\n")
print("Nossa análise ponderou os critérios mais relevantes para o negócio, atribuindo maior peso ao faturamento, seguido por avaliação dos clientes, custo de frete e estabilidade de vendas ao longo do ano.")
print("Lojas que apresentam alta sazonalidade (ou seja, dependem de picos de vendas em poucos meses) são mais arriscadas para manter a longo prazo, pois podem sofrer oscilações importantes no fluxo de caixa.")
print(f"Com base nessa análise integrada, a loja recomendada para venda é a **{loja_para_vender}**.")
print("Essa decisão considera não só o desempenho atual mas também a regularidade das vendas (baixa sazonalidade), a satisfação dos clientes e os custos logísticos.")
print("\n**Resumo dos motivos para a escolha:**")
print("- Menor pontuação integrada considerando todos os critérios.")
print("- Faturamento relativamente inferior ou alto grau de sazonalidade (instabilidade nas vendas).")
print("- Avaliações dos clientes e frete médio menos competitivos em relação às outras lojas.\n")
print("**Recomendação:** Mantenha as lojas com maior faturamento, boa avaliação e vendas estáveis ao longo do ano, garantindo assim a saúde financeira e reputacional do grupo no médio e longo prazo.")

# --- MELHORIAS PARA AS LOJAS MANTIDAS ---
print("\n## Sugestões de Melhoria para as Lojas Mantidas:")

for idx in idx_mantidas:
    print(f"\n### {labels_lojas[idx]}")
    print("- **Avaliação dos clientes:**")
    if avaliacoes[idx] < 4:
        print("  - Invista em treinamento da equipe para melhorar o atendimento e implemente pesquisas de satisfação pós-venda.")
    else:
        print("  - Continue monitorando e incentivando feedbacks positivos dos clientes.")
    print("- **Frete:**")
    if fretes[idx] > np.mean(fretes):
        print("  - Negocie melhores condições com transportadoras ou ofereça promoções de frete grátis para compras acima de determinado valor.")
    else:
        print("  - Mantenha o controle dos custos logísticos para seguir competitivo.")
    print("- **Faturamento e Sazonalidade:**")
    if sazonalidade[idx] > np.mean(sazonalidade):
        print("  - Crie campanhas promocionais fora dos meses de pico para reduzir dependência da sazonalidade.")
    else:
        print("  - Aproveite a regularidade e invista em fidelização de clientes e expansão do mix de produtos.")
    print("- **Mix de Produtos:**")
    print("  - Analise os produtos menos vendidos e considere substituí-los por itens mais demandados ou novidades do mercado.")
    print("- **Marketing:**")
    print("  - Invista em campanhas digitais e parcerias locais para aumentar o alcance da loja.")

print("\nEstamos à disposição para detalhar qualquer aspecto adicional ou realizar projeções específicas, caso necessário.")
```
